```python
from scipy import integrate
from sklearn.metrics import mean_squared_error
import math
```

```python
def calc(y_:np.ndarray):
    best_mu,best_sig,best_preds = -1,-1,[]
    minsse = float('inf')
    rg = [
        [-100,2.5], # 2
        [2.5,3.5],
        [3.5,4.5],
        [4.5,5.5],
        [5.5,6.5],
        [6.5,100],
    ]
    for mu in np.arange(2.5,4.5,0.1):
        for sig in np.arange(1,6.2,0.1):
            def func(x):
                return 1/(sig*math.sqrt(2*math.pi))*math.exp(-(x-mu)**2/(2*sig**2))
            preds = []
            for i in range(6):
                val,_ = integrate.quad(func,rg[i][0],rg[i][1])
                # print(val)
                preds.append(val)
            # print(len(preds))
            err = mean_squared_error(y_,np.array(preds))
            if np.sum(preds)<0.1:
                print(preds,sig,mu)
                assert(0)
            if err < minsse:
                minsse,best_mu,best_sig = err,mu,sig
                best_preds = preds
    return best_preds,best_mu,best_sig
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
from pylab import mpl
import lightgbm as lgb
import xgboost as xgb
from sklearn import preprocessing
from sklearn import svm 
from sklearn.model_selection import KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from sklearn.decomposition import PCA
import matplotlib
import matplotlib.pyplot as plt 
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
pd.set_option('max_columns',1000)
pd.set_option('max_row',300)
np.set_printoptions(threshold=np.inf)
```

```python
K = 8
kf = KFold(8,shuffle=True,random_state=520,)

#poisson regression

lgb_params = {
    'learning_rate': 0.01,
    'boosting_type': 'gbdt',
    'objective': 'poisson',
    'bagging_fraction': 0.8,
    'bagging_freq':1,
    'num_leaves': 80,
    'colsample_bytree': 0.6,
    'min_data': 200,
    'min_hessian': 1,   
    'verbose': -1
}
def evalerror(pred,df):
    label = df.get_label().copy()
    score = mean_squared_error(label,pred)*0.5
    return ('0.5 mse',np.sqrt(score),False)
```

```python
df.columns
```

```python
used_feats = [
    'score1','score2','freq_word','nvowel','relation','smlt'
]
df_train_X = df[used_feats]
df_train_y1 = pd.DataFrame(labels[:,0],columns=['sigma'])
df_train_y2 = pd.DataFrame(labels[:,1],columns=['mu'])
df_test_X = pd.DataFrame([[4,4,0.00025,2,1,3]],columns=used_feats,index=np.arange(1))
```

```python
train_preds_lgb1 = np.zeros(df_train_X.shape[0])
train_preds_lgb2 = np.zeros(df_train_X.shape[0])
test_preds_lgb1 = np.zeros((df_test_X.shape[0],K))
test_preds_lgb2 = np.zeros((df_test_X.shape[0],K))

for i,(train_idx,val_idx) in enumerate(kf.split(df_train_X)):
    print('\n {}...'.format(i))
    train_feat1 = df_train_X.iloc[train_idx]
    train_feat2 = df_train_X.iloc[val_idx]
    train_target1 = df_train_y1.iloc[train_idx]
    train_target2 = df_train_y1.iloc[val_idx]


    print('lightgbm for sigma')
    lgb_train1 = lgb.Dataset(train_feat1.values,train_target1.values)
    lgb_train2 = lgb.Dataset(train_feat2.values,train_target2.values)

    gbm = lgb.train(
        lgb_params,
        lgb_train1,
        num_boost_round=20000,
        valid_sets=lgb_train2,
        verbose_eval=500,
        feval=evalerror,
        early_stopping_rounds=200    
    )
    train_preds_lgb1[val_idx] += gbm.predict(train_feat2)
    test_preds_lgb1[:,i] = gbm.predict(df_test_X)
    
    train_target1 = df_train_y2.iloc[train_idx]
    train_target2 = df_train_y2.iloc[val_idx]


    print('lightgbm for mu')
    lgb_train1 = lgb.Dataset(train_feat1.values,train_target1.values)
    lgb_train2 = lgb.Dataset(train_feat2.values,train_target2.values)

    gbm = lgb.train(
        lgb_params,
        lgb_train1,
        num_boost_round=20000,
        valid_sets=lgb_train2,
        verbose_eval=500,
        feval=evalerror,
        early_stopping_rounds=200    
    )
    train_preds_lgb2[val_idx] += gbm.predict(train_feat2)
    test_preds_lgb2[:,i] = gbm.predict(df_test_X)
    print('\n')
```

```python
def predict6p(sig:float,mu:float):
    rg = [
        [-100,1.5], # 2
        [1.5,2.5],
        [2.5,3.5],
        [3.5,4.5],
        [4.5,5.5],
        [5.5,6.5],
        [6.5,100]
    ]
    def func(x):
        return 1/(sig*math.sqrt(2*math.pi))*math.exp(-(x-mu)**2/(2*sig**2))
    preds = []
    for i in range(6):
        val,_ = integrate.quad(func,rg[i][0],rg[i][1])
        preds.append(100*val)
        # print(val)
    return np.array(preds,dtype=int)
```
